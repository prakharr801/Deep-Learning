```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/Student_Performance.csv")
```

```python
df.head()
```

```python
plt.scatter(df.cgpa,df.package)
plt.xlabel("cgpa")
plt.ylabel("package in lakhs")
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
##already imported train_test_split [1 2 m 5 l 7 7 m]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
lnr = LinearRegression()
```

```python
lnr.fit(X_train,Y_train)
```

```python
lnr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
PLEASE_ENTER_YOUR_CGPA = 10
```

```python
lnr.predict(np.array([[PLEASE_ENTER_YOUR_CGPA]]))
```

```python
plt.scatter(df.cgpa,df.package)
plt.plot(X_train,lnr.predict(X_train),color="red")
plt.plot(X_test,lnr.predict(X_test),color="green")
plt.xlabel("cgpa")
plt.ylabel("package in lakhs")
```
